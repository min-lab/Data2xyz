```python
from bokeh.io import output_notebook, push_notebook, show
from bokeh.models import Span
from bokeh.plotting import figure

import pandas as pd
import glob
from pyexcelerate import Workbook
from pathlib import Path
```

```python
o_dir = './excel/'
skip_file = ['0504sitrightdrinkh2o.csv','0504stand_3min.csv']
```

```python
def cut_data(data_df):
    fragment = []
    silence = False
    data_filtered = (data_df["ax"] < 0.3) & (data_df["ax"] > -0.3)
    cnt = 0
    for idx in range(len(data_filtered)):
        if data_filtered[idx]:
            cnt = cnt + 1
        else:
            cnt = 0
            if silence:
                fragment.append(idx - 1)
                silence = False
        if cnt > 300:
            silence = True
            if idx == len(data_filtered)-1:
                fragment.append(idx)
    silence = False
    data_filtered = (data_df["ax"] < 0.3) & (data_df["ax"] > -0.3)
    cnt = 0
    for idx in reversed(range(len(data_filtered))):
        if data_filtered[idx]:
            cnt = cnt + 1
        else:
            cnt = 0
            if silence:
                fragment.append(idx + 1)
                silence = False
        if cnt > 300:
            silence = True
            if idx == 0:
                fragment.append(idx)
    fragment.sort()
    return fragment
```

```python
def show_data(data_df, fragment):
    output_notebook()
    p = figure(
        title="origin",
        x_axis_label="time",
        y_axis_label="ax",
        plot_width=1500,
        plot_height=800,
    )
    p.line(data_df.time, data_df.ax, legend_label="Amplitude", line_width=2)
    for idx in fragment:
        line = data_df.time[idx]
        vline = Span(
            location=line,
            dimension="height",
            line_color="red",
            line_dash="dashed",
            line_width=3,
        )
        p.add_layout(vline)
    handle = show(p, notebook_handle=True)
    push_notebook(handle=handle)
```

```python
Path(o_dir).mkdir(parents=True, exist_ok=True)
for file in glob.glob('*.csv'):
    data = pd.read_csv(file).drop(columns=['Unnamed: 7'])
    file_cnt = 0
    wb = Workbook()
    if file in skip_file:
        data = data.drop_duplicates(subset ="time", keep=False) #remove duplicate
        print("Write" + o_dir + '/' + file[:-4] + ".xlsx, sheet_name:" + "raw")
        output_data = [
                data.columns.tolist(),
            ] + data.values.tolist()
        wb.new_sheet("raw", data=output_data)
        wb.save(o_dir + file[:-4] + ".xlsx")
    else:
        fragment = cut_data(data)
        show_data(data, fragment)
        data = data.drop_duplicates(subset ="time", keep=False) #remove duplicate
        for idx in range(0, len(fragment) - 3, 2):
            file_cnt = file_cnt + 1
            print("Write" + o_dir + file[:-4] + ".xlsx, sheet_name:" + "raw_" + str(file_cnt))
            file_idx_s = fragment[idx]
            file_idx_e = fragment[idx + 3]
            output_data = [
                data[file_idx_s:file_idx_e].columns.tolist(),
            ] + data[file_idx_s:file_idx_e].values.tolist()
            wb.new_sheet("raw_" + str(file_cnt), data=output_data)
            wb.save(o_dir + file[:-4] + ".xlsx")
            
```
