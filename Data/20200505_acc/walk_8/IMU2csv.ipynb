```python
from bokeh.io import output_notebook, push_notebook, show, export_png
from bokeh.models import Span
from bokeh.plotting import figure
import pandas as pd
from pathlib import Path
```

```python
i_file = "imu0502__20.txt"
o_dir = "../"
image_name = '加速規數據.png'
csv_name = 'accelerometer.csv'
```

```python
data = []
with open(i_file) as f:
    lines = list(f)[26:-3]
    for line in lines:

        st = line.split(",")
        data.append(st)
        
        
data_df = pd.DataFrame(data, columns=["x", "y", "z", "time"])
data_df[["x", "y", "z"]] = data_df[["x", "y", "z"]].astype(float)
data_df["time"] = data_df["time"].astype(int)
data_df["duration"] = data_df.time.cumsum()
```

```python
output_notebook()
p = figure(
    title="加速規數據",
    x_axis_label="時間(ms)",
    y_axis_label="加速度(m/s)",
    plot_width=1200,
    plot_height=800,
)
p.line(data_df.duration, data_df.y, legend_label="y軸加速度", line_width=2)
start = 3000
vline_start = Span(
        location=start,
        dimension="height",
        line_color="red",
        line_dash="dashed",
        line_width=3,
)
p.add_layout(vline_start)

stop = 11000
vline_stop = Span(
        location=stop,
        dimension="height",
        line_color="red",
        line_dash="dashed",
        line_width=3,
)
p.add_layout(vline_stop)
handle = show(p, notebook_handle=True)
push_notebook(handle=handle)
data_df = data_df[(data_df.duration.values >= 3000) & (data_df.duration.values <= 11000)]
```

```python
Path(o_dir).mkdir(parents=True, exist_ok=True)
export_png(p, filename= o_dir + image_name)
data_df[['x','y','z','duration']].to_csv(o_dir + csv_name ,index=False)
```
